```python
import csv
import numpy as np
players = []
stats = []
seasons = []
teams = []
                
```

```python
with open('all_seasons.csv', newline='') as csvFile:
    seasonReader = csv.DictReader(csvFile)
    seasonID = 200000

    unique_seasons = []
    for row in seasonReader:
        if not(row['season'] in unique_seasons):
            unique_seasons.append(row['season'])
    
    for season in unique_seasons:
        seasonDict = {'SeasonID': seasonID, 'Season': season}
        seasons.append(seasonDict)
        seasonID += 1
```

```python
with open('seasons.csv', mode='w') as seasons_csv:
    season_headers = ['SeasonID', 'Season']
    
    season_writer = csv.DictWriter(seasons_csv, fieldnames = season_headers)
    season_writer.writeheader()
    for item in seasons:
        season_writer.writerow(item)
```

```python
with open('all_seasons.csv', newline='') as csvFile:
    playerReader = csv.DictReader(csvFile)
    playerID = 300000

    for row in playerReader:
        if not (row['draft_year'] == 'Undrafted'):
            draft_decade = str(int(np.floor(int(row['draft_year'])/10) * 10)) + 's'
        else:
            draft_decade = "Undrafted"
        playerDict = {'PlayerID': playerID, 'Name': row['player_name'], 'College': row['college'] , 'Draft_Year': row['draft_year'] , 'Age': row['age'], 'Height': row['player_height'], 'Weight':row['player_weight'], 'Draft_Decade': draft_decade}
        players.append(playerDict)
        playerID += 1
```

```python
with open('players.csv', mode='w') as players_csv:
    player_headers = ['PlayerID', 'Name', 'College', 'Draft_Year', 'Age', 'Height', 'Weight', 'Draft_Decade']
    
    player_writer = csv.DictWriter(players_csv, fieldnames = player_headers)
    player_writer.writeheader()
    for item in players:
        player_writer.writerow(item)
```

```python
with open('all_seasons.csv', newline='') as csvFile:
    teamID = 400000
    teamReader = csv.DictReader(csvFile)
    unique_teams = []
    for row in teamReader:
        if not(row['team_abbreviation'] in unique_teams):
            unique_teams.append(row['team_abbreviation'])
    
    for team in unique_teams:
        teamDict = {'TeamID': teamID, 'team_abbreviation': team}
        teams.append(teamDict)
        teamID += 1
```

```python
with open('teams.csv', mode='w') as team_csv:
    team_headers = ['TeamID', 'team_abbreviation']
    
    team_writer = csv.DictWriter(team_csv, fieldnames = team_headers)
    team_writer.writeheader()
    for item in teams:
        team_writer.writerow(item)
```

```python
with open('all_seasons.csv', newline='') as csvFile:
    statsReader = csv.DictReader(csvFile)
    statID = 100000
    seasonID = 0
    playerID = 300000
    teamID = 0
    for row in statsReader:
        for season in seasons:
            if(season['Season'] == row['season']):
                seasonID = season['SeasonID']
        for team in teams:
            if(team['team_abbreviation'] == row['team_abbreviation']):
                teamID = team['TeamID']
        statsDict = {'StatsID': statID, 'SeasonID': seasonID, 'PlayerID': playerID, 'TeamID': teamID, 
                        'AvgPoints': row['pts'], 'AvgAssists': row['reb'], 'AvgRebounds': row['ast'], 'OffRebounds': row['oreb_pct'], 'DefRebounds': row['dreb_pct'],                          'GP': row['gp'], 'Net_Rating': row['net_rating'], 'TeamPlays': row['usg_pct'], 'AssistPct': row['ast_pct'], 'ShootingEff': row['ts_pct']
                    }
        stats.append(statsDict)
        statID += 1
        playerID += 1 # cause parallel
    
```

```python
with open('stats.csv', mode='w') as stats_csv:
    stats_headers = ['StatsID', 'SeasonID', 'PlayerID', 'TeamID','AvgPoints', 'AvgAssists', 'AvgRebounds', 'OffRebounds', 'DefRebounds', 'GP', 'Net_Rating', 'TeamPlays', 'AssistPct', 'ShootingEff']
    
    stats_writer = csv.DictWriter(stats_csv, fieldnames = stats_headers)
    stats_writer.writeheader()
    for item in stats:
        stats_writer.writerow(item)
```
